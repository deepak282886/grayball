```python
import os 
# Find all table elements 
def get_scorecard(soup, folder):
  tables = soup.find_all('table')

  # Iterate through tables and extract data to DataFrames
  dataframes = []
  for table in tables:
      rows = []
      for row in table.find_all("tr"):
          cells = []
          for cell in row.find_all(["td", "th"]):
              cells.append(cell.get_text().strip())
          rows.append(cells)
      
      df = pd.DataFrame(rows)
      dataframes.append(df)

  header_row = 0
  count = 0
  os.makedirs(f"match_data/{folder}/", exist_ok=True)
  for df_x in dataframes:
    df = df_x.dropna()
    df.columns = df.iloc[header_row]
    final_df = df.reset_index(drop=True)
    final_df.to_csv(f"match_data/{folder}/match_df_{count}.csv", index=False)
    count += 1

def get_match_report(soup, folder):
  os.makedirs(f"match_data/{folder}/", exist_ok=True)
  text = str(soup).split('"articleBody":"')[1].split('","articleSection')[0]
  with open(f"match_data/{folder}/match_report.txt", "w") as f:
    f.write(text)
```

```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd

for first in list_1:
  try:
    response1 = requests.get(f"https://www.espncricinfo.com{first}")
    soup1 = BeautifulSoup(response1.text, 'lxml')
    first_list = []
    for link in soup1.find_all('a'):
      hre = link.get('href')
      if "https://www.espncricinfo.com/series/" in hre:
        hre2 = "/".join(hre.split("/")[:-1])
        first_list.append(hre2)
    first_list = list(set(first_list))
    second_list = []
    if len(first_list):
      for second in first_list:
        series = second.split("/")[-1].split("-")[-1] + "/match-schedule-fixtures-and-results"
        response2 = requests.get(second)
        soup2 = BeautifulSoup(response2.text, 'lxml')
        for link in soup2.find_all('a'):
          hre = link.get('href')
          if series in hre and "espncric" not in hre:
            second_list.append(f"https://www.espncricinfo.com{hre}")
    
      second_list = list(set(second_list))
    third_list= []
    if len(second_list):
      for third in second_list:
        series = third.split("/")[-2].split("-")[-1]
        response2 = requests.get(third)
        soup2 = BeautifulSoup(response2.text, 'lxml')
        for link in soup2.find_all('a'):
          hre = link.get('href')
          if series in hre and "full-scorecard" in hre:
            third_list.append(f"https://www.espncricinfo.com/{hre}")
      third_list = list(set(third_list))
    
    page_list = ["full-scorecard", "match-report"]
    if len(third_list):
      for fourth in third_list:
        match_number = fourth.split("/")[-2].split("-")[-1]
        for k in page_list:
          new_fourth = fourth.replace("full-scorecard", k)
          response2 = requests.get(new_fourth)
          soup2 = BeautifulSoup(response2.text, 'lxml')
          if k == "full-scorecard":
            get_scorecard(soup2, match_number)
          if k == "match-report":
            get_match_report(soup2, match_number)
  except Exception as e:
    print(e)
    pass
```
