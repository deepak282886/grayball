```python
import sqlite3

# Assuming you are using SQLite as the database
conn = sqlite3.connect("cricket.db")
cursor = conn.cursor()
```

```python
def team_info(team):
  cols = ['match_id', 'team', 'over_number', 'powerplay_type', 'powerplay_number', 'ball_number', 'batter',
          'bowler', 'non_striker', 'runs_batter', 'extras_total', 'target_overs', 'target_runs', 'wickets_player_out',
          'wickets_kind', 'fielder1_wicket', 'fielder2_wicket', 'fielder3_wicket', 'extras_type1', 'extras_type2', 'runs_extras1',
          'runs_extras2', 'match_id_2', 'balls_per_over', 'city', 'dates', 'event_name', 'event_match_number', 'gender', 'match_type',
          'match_type_number', 'official_match_referees', 'official_reserve_umpires', 'official_tv_umpires', 'official_umpires',
          'outcome_winner', 'outcome_wickets', 'outcome_runs', 'overs', 'player_of_match', 'players_team1', 'players_team2', 'season',
          'team_type', 'teams', 'toss_decision', 'toss_winner', 'venue', 'home_team', 'away_team']

  query = f"""
      SELECT *
      FROM ball_by_ball_data b
      LEFT JOIN match_info m ON b.match_id = m.match_id
      WHERE '{team}' IN (m.home_team, m.away_team);
  """
  team_df = pd.read_sql_query(query, conn)
  team_df.columns = cols

  return team_df
```

```python
big_df_list = []
for i in nation_list:
  team_df = team_info(i)
  bat_df = team_df[team_df["team"] == i]
  bowl_df = team_df[team_df["team"] != i]
  bats = list(bat_df["batter"].unique())
  bats1 = list(bat_df["non_striker"].unique())
  balls = list(bowl_df["bowler"].unique())
  player_list = []
  player_list.extend(bats)
  player_list.extend(bats1)
  player_list.extend(balls)
  player_list = list(set(player_list))
  if len(player_list):
    dt = pd.DataFrame(player_list)
    dt.columns = ["players"]
    dt["nation"] = i
    big_df_list.append(dt)
  else:
    pass
```

```python
big_df = pd.concat(big_df_list)
```

```python
sdf_list = []
for name, sdf in big_df.iterrows():
    nation = sdf['nation']
    player = sdf['players']
    ddf = df_names_full[df_names_full["nation"] == nation]

    if len(ddf[ddf["name"] == player]):
      tdf = ddf[ddf["name"] == player]
      sdf["name"] = tdf["name"].values[0]
      sdf["full_names"] = tdf["full_names"].values[0]
      sdf["r1"] = tdf["r1"].values[0]
      sdf["r2"] = tdf["r2"].values[0]
      sdf_list.append(sdf)
    elif len(ddf[ddf["full_names"] == player]):
      tdf = ddf[ddf["full_names"] == player]
      sdf["name"] = tdf["name"].values[0]
      sdf["full_names"] = tdf["full_names"].values[0]
      sdf["r1"] = tdf["r1"].values[0]
      sdf["r2"] = tdf["r2"].values[0]
      sdf_list.append(sdf)
    elif len(ddf[ddf["r1"] == player]):
      tdf = ddf[ddf["r1"] == player]
      sdf["name"] = tdf["name"].values[0]
      sdf["full_names"] = tdf["full_names"].values[0]
      sdf["r1"] = tdf["r1"].values[0]
      sdf["r2"] = tdf["r2"].values[0]
      sdf_list.append(sdf)
    elif len(ddf[ddf["r2"] == player]):
      tdf = ddf[ddf["r2"] == player]
      sdf["name"] = tdf["name"].values[0]
      sdf["full_names"] = tdf["full_names"].values[0]
      sdf["r1"] = tdf["r1"].values[0]
      sdf["r2"] = tdf["r2"].values[0]
      sdf_list.append(sdf)
```
